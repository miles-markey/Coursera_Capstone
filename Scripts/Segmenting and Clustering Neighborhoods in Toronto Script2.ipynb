The following cell performs all the functionality of the first script of this project to give an updated dataframe to work with

```python
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfs = pd.read_html(url)
df = dfs[0]
#Drop all rows where the borough is "Not assigned"
newDf1 = df[df.Borough != 'Not assigned']
```

The next step is to latitude and longitude of each neighborhood into the dataframe. The geocoder package was not performing well so the given csv file was used for the latitude/longitude data. This file was read into a dataframe called zipData, which was then used to append the original dataframe. df.join is used to merge the two dataframes

```python
zipData = pd.read_csv('Geospatial_Coordinates.csv')

newDf1.join(zipData.set_index('Postal Code'), on='Postal Code')
```
